```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get data from csv file
train_set1 = pd.read_csv('Dataset_2_train.csv',delimiter =',', header=None)
val_set1 = pd.read_csv('Dataset_2_valid.csv', delimiter =',', header=None)
test_set1 = pd.read_csv('Dataset_2_test.csv',delimiter =',', header=None)

# To save w values
w = [0.0,0.0]
# Saving x and y as vectors 

x_train = train_set1[0]
x_val = val_set1[0]
x_test = test_set1[0]
y_train = train_set1[1]
y_val = val_set1[1]
y_test = test_set1[1]

```

```python
# plot the data to see how it looks like
plt.figure(0)
plt.plot(x_train,y_train, 'ko', label ='h')
plt.xlabel("x train")
plt.ylabel("y train")
plt.title("Training Set")
plt.figure(1)
plt.plot(x_val,y_val, 'bo')
plt.xlabel("x validation")
plt.ylabel("y validation")
plt.title("Validation Set")
plt.show()
```

```python

# Defining functions for gradient descent. Start by defining global variables.
step = 10**(-6)
train_mse = []
validation_mse = []
epoch = 0
# print (step)
def mse(y,y_hat):
    return sum([(y_hat[i] - y[i]) ** 2 for i in range(len(y))])/len(y)

def mseList(x,y):
    tempX = []
    tempY = []
    output = []
    for i in range(len(np.array(x))):
        tempX.append(x[i])
        tempY.append(y[i])
        updateParam(x[i], y[i])
        yExpected = [y_hat(j) for j in tempX] 
        output.append(mse(tempY, yExpected))
    return output

def updateData(x,y):
    for i in range(len(np.array(x))):
        updateParam(x[i], y[i])

def updateParam(x,y):
        w[0] = w[0] - (step * ((y_hat(x)) - y))
        w[1] = w[1] - (step * ((y_hat(x)) - y) * x)

def y_hat(x):
    return w[0] + w[1]*x

def shuffle_data(x,y):
    shuf = np.random.permutation(len(y))
    return x[shuf],y[shuf]

values = mseList(x_val,y_val)
train = []
valid = []

w_next = [0.1, 0.1]

numEpoch = 0
#for i in range(numEpoch):

#epsilon = 0.0001
while np.linalg.norm(np.array(w_next)-np.array(w))>0.0001:
    #print(w, w_next)
    #print(np.linalg.norm(np.array(w_next)-np.array(w)))
    w_next[0] = w[0]
    w_next[1] = w[1]
    x_train, y_train = shuffle_data(x_train,y_train)
    numEpoch = numEpoch +1
    updateData(x_train, y_train)
    updateData(x_val, y_val)
    train.append(mse(y_train,[y_hat(x_train[i]) for i in range(len(x_train))]))
    valid.append(mse(y_val,[y_hat(x_val[i]) for i in range(len(x_val))]))
    #updateParam(x_train[numEpoch], y_train[numEpoch])
    

plt.figure(2)
plt.plot(x_train, y_train, 'go', label = 'Training')
plt.plot(x_val, y_val, 'ko', label = 'Validating')
plt.plot(x_val, [y_hat(x_val[i]) for i in range(len(x_val))],'r', label='fitting curve')
plt.legend(loc='best')
plt.show()
print('Value of w0, w1 is ' +str(w)+ ' and MSE is '+ str(valid[-1])+ ' after '+str(numEpoch)+' epochs')
print('MSE for each epoch is shown below')
#print(valid)
```
